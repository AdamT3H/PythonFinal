```python
class Animal:
    def __init__(self, type, name, age):
        self._type = type
        self._name = name
        self._age = age

    def __str__(self) -> str:
        return f"Це є {self._name} і він роду {self._type}, його вік {self._age} років, троха старе."



class Mammal(Animal):
    def __init__(self, name, age, habitat):
        super().__init__("Mammal", name, age)
        self.habitat = habitat

    def info(self):
        return f"{super().__str__()} Живе у {self.habitat}."

    def sleep(self):
        return f"{self._name} спить чомусь("
    
class Bird(Animal):
    def __init__(self, name, age, habitat):
        super().__init__("Bird", name, age)
        self.habitat = habitat

    def info(self):
        return f"{super().__str__()} Живе у {self.habitat}."
    
    def fly(self):
        return f"{self._name} летить на роботу("

class Reptile(Animal):
    def __init__(self, name, age, habitat):
        super().__init__("Reptile", name, age, habitat)
        self.habitat = habitat

    def info(self):
        return f"{super().__str__()} Живе у {self.habitat}."
    
    def creeps(self):
        return f"{self._name} повзе з роботи на другу роботу на нічну зміну((((("
    
class Zoo():
    def __init__(self):
        self.animals = []
        self.type_of_animals = { 
            "reptile":[], 
            "bird":[], 
            "mammal":[]
        }

    def list_of_animals_by_type(self):
        list_of_animals_by_type = []
        for category, animals in self.type_of_animals.items():
            list_of_animals_by_type.append(f"{category}:")
            if animals:
                for animal in animals:
                    list_of_animals_by_type.append(f" - {animal._name}")
            else:
                list_of_animals_by_type.append(" - віс тварини цієї категорії втекли(")
        return list_of_animals_by_type

    def add_animal(self, animal):
        self.animals.append(animal)
        if (animal._type == "Mammal"):
            self.type_of_animals["mammal"].append(animal)
        elif (animal._type == "Bird"):
            self.type_of_animals["bird"].append(animal)
        elif (animal._type == "Reptile"):
            self.type_of_animals["reptile"].append(animal)


    def list_of_animals(self):
        print("Список енімалс: ")
        for animal in self.animals:
            print(animal._name)

    def find_oldest(self):
        print("Найстарший серед Mammals: ")
        mamal_oldest = []
        for mammal in self.type_of_animals["mammal"]:
            mamal_oldest.append(mammal._age)
        mamal_oldest.sort()
        print(f'Самий дід серед Mammals є {mamal_oldest[-1]}')


Cat = Mammal("Кіт", 98, "Місто тернопіль")
print(Cat.info())
print(Cat.sleep())

Pig = Mammal("Свинка", 18, "Вашингтон")
print(Pig.info())
print(Pig.sleep())

Sparow = Bird("Горобець", 12, "ППО")
print(Sparow.info())
print(Sparow.fly())


Zoo = Zoo()
Zoo.add_animal(Cat)
Zoo.add_animal(Pig)
Zoo.add_animal(Sparow)
Zoo.list_of_animals()
Zoo.find_oldest()
Zoo.list_of_animals_by_type()
```
