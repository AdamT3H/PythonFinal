```python
import threading
from uu import Error
from datetime import datetime
import os

class LoggingError(Exception):
    def __init__(self):
        self.message = "Not good, you have 3 errors!!! very bed!"

    def __str__(self):
        return self.message


class Logger:
    def __init__(self, file = "file.log"):
        self.file = file

    errorCount = 0

    try:
        def _writeLog(self, level, message):
            with open(self.file, "a") as fileKyrache:
                fileKyrache.write(f"{level} : {message}\n")
    except Error: # тут можна замінити Error на любу помилку, як я розумію, ну  головне шо знаю як трай писати 
        print("You have some problemas")
        errorCount += 1

    if errorCount == 3:
        raise LoggingError

    def info(self, message):
        self._writeLog("info", message)

    def warning(self, message):
        self._writeLog("warning", message)

    def error(self, message):
        self._writeLog("error", message)

    def critical(self, message):
        self._writeLog("critical", message)

    def archiveOldLogs(self):
        try:
            archiveForLogs = f"logs_archive_{datetime.now().strftime('%Y.%m.%d_%H:%M:%S')}.log"
            
            with open(archiveForLogs, 'w') as archive:

                with open(self.file, 'r') as logs:
                    archive.write(logs.read())
            print(f"Archived log file to {archiveForLogs}")
            os.remove(self.file)

        except FileNotFoundError:
            print(f"Log file {self.file} is not here")


def runDaily():
    logger = Logger()
    logger.info("not goooood!!! very bed!")
    logger.warning("not good!!! very bed!")
    logger.error("not gd!!! very bed!")
    logger.critical("not!!! very bed!")
    threading.Timer(86400, runDaily).start() # тут використав цей метод, томушо з доступних в інтернеті він здався найпростішим та найадекватнішим.
    logger.archiveOldLogs()



if __name__ == "__main__":
    runDaily()
```
